### Libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

---

### 1. In the Drive folder you have a dataset 'Diabetes_data.csv', read the dataset into Pandas.

```python
df = pd.read_csv('Diabetes_data.csv')
df
```

---

### 2. The data is slopily gathered and you need to rename, correct, and handle missing values. Check to see if any of the columns have null/missing values and determine the best strategy for cleaning up the data (including not doing anything).

```python
df.to_html('dataset.html')
```

```python
df = df.rename(columns={'Kolku dolgo e dijabeticar': 'DiabetesForHowLong(years)', 'Kolku mu e tromesechen (hemoglobin A1C) she4er': 'ThreeMonthsHemoglobin'})
```

```python
df = df.rename(columns={'Dali e na insulin ili na tabli?': 'InsulinOrPill', 'Glukoza (momentalna) vo krv': 'PresentGlucose','Dali ima dijabetesna retinopatija?': 'DiabeticRetinopathy'})

```

```python
df['NumberOfPatient'] = df.reset_index().index
df
```

---

The Data is clean and we can continue with proccesing


### 3. Find the patient which currently has the highest glucose level. Sketch (on a plot of your choice) all of the patients glucose levels and describe the plot. Draw a conclusion on how the data behaves.

```python
max_glucose_patient = df.loc[df['PresentGlucose'].idxmax()]
print('The patient with the highest glucose level is:\n', max_glucose_patient)
```

```python
plt.plot(df['PresentGlucose'], 'o')
plt.xlabel('Patient')
plt.ylabel('Glucose Level')
plt.title('Glucose Levels of Patients')
plt.show()
```

We can confirm that the information about the patient with highest glucose is correct. We can conclude that the glucose levels of these patients are relatively close together, with some variation between patients. There is no clear trend or pattern in the data.

### 4. Find the average number of hemoglobin A1C levels of patients per length of disease. Which duration of the disease has the highest average glucose levels? Sketch the data on a plot of your choice to best describe the data. Write a sentence or two describing how the data behaves.

```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
average_hemoglobin = df.groupby('DiabetesForHowLong(years)')['ThreeMonthsHemoglobin'].mean()


plt.plot(average_hemoglobin, 'o-')
plt.xlabel('Length of Disease (years)')
plt.ylabel('Average Hemoglobin A1C Level')
plt.title('Average Hemoglobin A1C Levels per Length of Disease')
plt.show()
```

```python
average_hemoglobin
```

```python

average_glucose = df.groupby('DiabetesForHowLong(years)')['PresentGlucose'].mean()


highest_glucose_group = average_glucose.idxmax()

print(f"The group with the highest average glucose levels has a duration of {highest_glucose_group} years of diabetes.")
```

---

### 5. Find the average duration of the disease for patients that are on insulin and patients that are on tablets. Compare the results. Which method that the patients use shows better results?

```python

insulin_duration = df[df['InsulinOrPill'] == 1]['DiabetesForHowLong(years)'].mean()
tablet_duration = df[df['InsulinOrPill'] == 0]['DiabetesForHowLong(years)'].mean()

print(f"The average duration of the disease for patients on insulin is {insulin_duration:.2f} years.")
print(f"The average duration of the disease for patients on tablets is {tablet_duration:.2f} years.")
```

---

There is not any significant difference in this dataset. The results are almost equal, so none of the method does not gives better results.

### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python

df['A1C_level'] = ['normal' if A1C < 5.7 else 'elevated' for A1C in df['ThreeMonthsHemoglobin']]
df['A1C_level']
```

---

```python
import matplotlib.pyplot as plt

plt.scatter(df['DiabetesForHowLong(years)'], df['ThreeMonthsHemoglobin'], c=df['A1C_level'].map({'normal': 'blue', 'elevated': 'red'}))


plt.xlabel('Years with diabetes')
plt.ylabel('Hemoglobin level')
plt.title('Hemoglobin level vs. Years with diabetes (Normal vs Elevated A1C level)')


plt.show()
```

### 7. How do factors influence the outcome whether a patient will have (or not) diabetic retinopathy? Analyse whether there is a correlation or not between diabetic retinopathy and other data in the set. Do this by separating the patients into 2 groups and analyzing the characteristics of the groups with respect to various information. Plot various data points if possible.

```python
import pandas as pd
import matplotlib.pyplot as plt


retinopathy_df = df[df['DiabeticRetinopathy'] == 1]
no_retinopathy_df = df[df['DiabeticRetinopathy'] == 0]

retinopathy_duration = retinopathy_df['DiabetesForHowLong(years)'].mean()
no_retinopathy_duration = no_retinopathy_df['DiabetesForHowLong(years)'].mean()
print('Average duration of diabetes for patients with diabetic retinopathy:', retinopathy_duration)
print('Average duration of diabetes for patients without diabetic retinopathy:', no_retinopathy_duration)

retinopathy_glucose = retinopathy_df['PresentGlucose'].mean()
no_retinopathy_glucose = no_retinopathy_df['PresentGlucose'].mean()
print('Average glucose level for patients with diabetic retinopathy:', retinopathy_glucose)
print('Average glucose level for patients without diabetic retinopathy:', no_retinopathy_glucose)

retinopathy_a1c = retinopathy_df['ThreeMonthsHemoglobin'].mean()
no_retinopathy_a1c = no_retinopathy_df['ThreeMonthsHemoglobin'].mean()
print('Average hemoglobin A1C level for patients with diabetic retinopathy:', retinopathy_a1c)
print('Average hemoglobin A1C level for patients without diabetic retinopathy:', no_retinopathy_a1c)



```

---

```python
plt.scatter(retinopathy_df['DiabetesForHowLong(years)'], retinopathy_df['PresentGlucose'], c='r', label='DiabeticRetinopathy')
plt.scatter(no_retinopathy_df['DiabetesForHowLong(years)'], no_retinopathy_df['PresentGlucose'], c='b', label='No Diabetic Retinopathy')
plt.xlabel('Years with Diabetes')
plt.ylabel('Glucose')
plt.legend()
plt.show()
```
